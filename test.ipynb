```python
import matplotlib.pyplot as plt

from tradinglib.plot import Plotter
from tradinglib.base import IO
```

```python
eurusd = IO.leer_csv_mt4(archivo='latest-MT4')
data = eurusd.iloc[-120:-36]
```

```python
plotter = Plotter(data_ohlc=data)
```

```python
fig, ax = plt.subplots()
plotter.plot_candlestick(ax=ax)
plotter.plot_line(ax=ax)
fig.set_size_inches(15,8)
plt.show()
```

```python
fig, ax = plt.subplots()
data['WMA'] = data.Close.ewm(span=10).mean()
data.Close.plot()
ax = data.WMA.plot(figsize=(15,8))
Plotter.plot_candlestick_timeindex(ax=ax, fuente=data, velas=40)
fig.set_size_inches(15,8)
plt.show()
```
